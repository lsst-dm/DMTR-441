### Test case LVV-T1755: Verify calculation of residual PSF ellipticity correlations for separations less than 1 arcmin

Verify that the DM system has provided the code to calculate the median residual PSF ellipticity correlations averaged over an arbitrary field of view for separations less than 1 arcmin, and assess whether it meets the requirement that it shall be no greater than **TE1 = 2.0e-5**.

**Discussion**: The TEx metrics are calculated and persisted per-visit, and thus reside in the `visit` table. To verify this requirement, we retrieve the measurements and perform additional calculations in the notebook. Eventually we plan to include these calculations, and related plots and metrics, in the `analysis_tools` package.

```python
testcoll = 'u/leget/LSSTCam/DM-46582/PSF_TEx_w37_run2'
butler = Butler('/repo/main', collections=[testcoll])
```

#### Retrieve the `visit_table` from the Butler:

```python
visitTable = butler.get("visit_table", instrument="LSSTCam")
```

The description of residual ellipticity correlation requirements in the [OSS](ls.st/oss) specifies that the requirements should be met for E_1, E_2, and E_X. The calculations for these have already been implemented, so that we must retrieve three separate columns from the `visit_table`: `psfTE1e1`, `psfTE1e2`, and `psfTE1ex`.

For each of these, we plot histograms of their distributions, and overlay a vertical line denoting the median value.

```python
te1_requirement = 2e-5

for i in ['1', '2', 'x']:
    plt.figure()
    plt.hist(visitTable[f'psfTE1e{i}'], bins=np.linspace(-1e-4, 1e-4, 100))
    plt.gca().xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
    plt.gca().ticklabel_format(axis='x', style='sci', scilimits=(0, 0))
    plt.xlabel(f"TE1e{i}")
    plt.ylabel("# visits")
    ylim = plt.ylim()
    plt.gca().fill_betweenx(ylim, -te1_requirement, te1_requirement, color='lightblue', alpha=0.5, label='Requirements TE1')
    medVal = np.nanmedian(visitTable[f'psfTE1e{i}'])
    plt.plot([medVal, medVal], ylim, 'k--', label="current")
    plt.ylim(ylim)
    plt.legend()
    plt.show()
```

## Results

We have demonstrated that software tools exist to calculate residual ellipticity correlations, and that those quantities can be easily compiled for analysis. We see in the figures above that the median residual ellipticity correlation on scales of less than 1 arcmin, TE1, is within the required threshold for all three ellipticity quantities. We deem the result of this test a **Pass**.
